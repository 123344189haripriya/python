```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv("/content/Car_Purchasing_Data.csv",encoding='ISO-8859-1')
df
```

```python
df.shape
```

```python
df = df.drop(['Customer Name','Customer e-mail','Country'],axis=1)
```

```python
df.head(7)
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,Y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
Y_train_prediction = reg.predict(X_train)
```

```python
from sklearn.metrics import r2_score
```

```python
print(f"Accuracy of multiple linear :{r2_score(Y_train,Y_train_prediction)} ")
```

```python
Y_test_prediction = reg.predict(X_test)
```

```python
print(f"Accuracy of multiple linear :{r2_score(Y_test,Y_test_prediction)} ")
```
